```python
import pandas as pd
import matplotlib.pyplot as plt
import os

files = [
    'datasets/shopping_dataset_2000.xlsx',
    'datasets/shopping_dataset_70000.xlsx',
    'datasets/shopping_dataset_120000.xlsx',
]

def analyze_column(series, col_name, file_name):
    series = series.dropna()
    print(f"\n--- {file_name} | Column: {col_name} ---")
    if pd.api.types.is_numeric_dtype(series):
        mean_val   = series.mean()
        median_val = series.median()
        mode_vals  = series.mode()
        print(f"Type: numeric")
        print(f"Mean: {mean_val:.4f}")
        print(f"Median: {median_val:.4f}")
    else:
        mode_vals = series.mode()
        print(f"Type: categorical")
    vc = series.value_counts()
    if vc.empty or vc.iloc[0] == 1:
        print("Все значения уникальны, моды нет")
    else:
        modes = series.mode().tolist()
        print(f"Mode(s): {modes}")
        top10 = vc.head(10)
        top10.plot(kind='bar', xlabel='')
        plt.title(f"Top 10 — {file_name} | {col_name}")
        plt.ylabel("Count")
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

for file_path in files:
    df = pd.read_excel(file_path)
    file_name = os.path.basename(file_path)
    for col in df.columns:
        analyze_column(df[col], col, file_name)
```
